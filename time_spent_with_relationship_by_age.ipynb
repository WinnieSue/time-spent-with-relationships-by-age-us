```python
import pandas as  pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
filename = ('/content/time-spent-with-relationships-by-age-us.csv')
df = pd.read_csv(filename)
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.columns = ['Entity', 'Code', 'Year', 'Alone',
              'Friends',
              'Children',
              'Family',
              'Partner',
              'Coworkers']
df.head()
```

Data looks clean. I am going to go ahead and visualize

```python
plt.figure(figsize=(12,12))
plt.style.use('seaborn')

# subplot 1
plt.subplot(3, 2, 1);
sns.lineplot(x = df['Year'], y = df['Alone'],color = 'blue')
plt.xlabel('Age')
plt.ylabel('Hours')
plt.title('Time spent alone, by age of respondent')
plt.xlim(left = 15, right = 80);
plt.ylim(bottom = 0, top = 500);
# subplot 2
plt.subplot(3,2,2);
sns.lineplot(x = df['Year'], y = df['Friends'],color = 'black')
plt.xlabel('Age')
plt.ylabel('Hours')
plt.title('Time spent with friends, by age of respondent')
plt.xlim(left = 15, right = 80);
plt.ylim(bottom = 0, top = 500);
# subplot 3
plt.subplot(3,2,3);
sns.lineplot(x = df['Year'], y = df['Children'],color = 'blue')
plt.xlabel('Age')
plt.ylabel('Hours')
plt.title('Time spent with Children, by age of respondent')
plt.xlim(left = 15, right = 80);
plt.ylim(bottom = 0, top = 500);
# subplot 4 
plt.subplot(3,2,4);
sns.lineplot(x = df['Year'], y = df['Family'],color = 'blue')
plt.xlabel('Age')
plt.ylabel('Hours')
plt.title('Time spent with siblings and family, by age of respondent')
plt.xlim(left = 15, right = 80);
plt.ylim(bottom = 0, top = 500);
# subplot 5
plt.subplot(3,2,5);
sns.lineplot(x = df['Year'], y = df['Partner'],color = 'black')
plt.xlabel('Age')
plt.ylabel('Hours')
plt.title('Time spent with partner, by age of respondent')
plt.xlim(left = 15, right = 80);
plt.ylim(bottom = 0, top = 500);
# subplot 6
plt.subplot(3,2,6);
sns.lineplot(x = df['Year'], y = df['Coworkers'],color = 'black')
plt.xlabel('Age')
plt.ylabel('Hours')
plt.title('Time spent with coworkers, by age of respondent')
plt.subplots_adjust(left=0.1,
                    bottom=0.1,
                    right=0.9,
                    top=0.9,
                    wspace=0.4,
                    hspace=0.4) # adding space between subplots
plt.xlim(left = 15, right = 80);
plt.ylim(bottom = 0, top = 500);
```

1. The Alone graph shows that as people get older, they spend more and more hours alone.
2. The Friends graph shows that while between 30 and eighty, people spend less hours with friends yet the younger people spend more hours with their friends.
3. The time people spend with Children is highest for people in the 30s and 40s. younger people and the elderly spend less time with children.
4. The time spent with siblings and family decreases as one ages.
5. Time spent with the partner increases with age.
6.The time spent with co workers varies with age.

```python
# I am going to plot all the subplots on one graph for comparison
sns.lineplot(x = df['Year'], y = df['Alone'],color = 'blue', label = 'Time Alone')
sns.lineplot(x = df['Year'], y = df['Friends'],color = 'orange', label = 'Time with Friends')
sns.lineplot(x = df['Year'], y = df['Children'],color = 'purple', label = 'Time with Children')
sns.lineplot(x = df['Year'], y = df['Family'],color = 'magenta', label = 'Time with family')
sns.lineplot(x = df['Year'], y = df['Partner'],color = 'black', label = 'Time ithe partner')
sns.lineplot(x = df['Year'], y = df['Coworkers'],color = 'green', label = 'Time with Coworkers')
plt.legend(loc= (1.02, 0))
plt.grid(axis = 'x')
plt.xlabel('Age')
plt.ylabel('Hours')
plt.title('Who Americans spend their time with by Age', fontsize = 13)
plt.xlim(left = 15, right = 80);
plt.ylim(bottom = 0, top = 500);
```
